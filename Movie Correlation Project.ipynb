```python
# import libraires

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
# bring in data

df = pd.read_csv(r'C:\Users\pc\Desktop\Power BI Dashboard\Movies2.csv')
```

```python
df
```

```python
df.head()
```

```python
# check for missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
    
```

```python
# lets establish the data types for columns

df.dtypes
```

```python
# change data type of some column
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['budget'].astype('int64')
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# drop duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# drop duplictaes in all the table
df.drop_duplicates()
```

```python
df
```

```python
# For correlation, budget is key to revenue
# company also has correlation with revenue
```

```python
# using scatterplot for revenue and budget

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget of films')
plt.show()
```

```python
# plotting budget and gross using the seaborn

sns.regplot(x='budget', y='gross', data=df)
```

```python
# We need to look at correlation in the data frame with numerical fields
```

```python
df.corr(method='pearson')  # there are other methods too, pearson is default
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix For Numbers')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
